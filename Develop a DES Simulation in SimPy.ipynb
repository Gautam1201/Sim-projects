## Simulating Pomodoro Technique

```python
import simpy
import random

time_available = 1200
session_per_book = 3
book_length = 120 #Average Book Length
study_time = 25 # Average Study time
relax_time = 10 # Average Time
no_of_books = 15

cur_book = 1
def study(env):
    global cur_book
    while True:
        book_time = 0
        session = 0
        for c in range(session_per_book):
            print(f"Session {session} begins, Start Book {cur_book} at {env.now}")
            book_time += (study_time - random.randint(0,10))
            yield env.timeout(study_time)
            print(f"Relaxing Phase begins at {env.now}")
            yield env.timeout(relax_time)
            print(f"Session {session} ends at {env.now}")
            session+=1
        print(f"Amount of book {cur_book} finished: {book_time/book_length}")
        cur_book += 1
        # Break of 45 minutes
        print("Long Break before next book")
        yield env.timeout(45)
        print("\n")
        
env = simpy.Environment()
env.process(study(env))
env.run(until=time_available)
```
